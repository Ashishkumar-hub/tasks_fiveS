PROBLEM 1: Display 'Manager' and 'Grand Total Sales', for sales done by the sales agents reporting these managers

```python
#solution code here
#joinig  sales team and pipeline data
merged_data = sales_teams.merge(sales_pipeline,on=["sales_agent"]) 
merged_data.head()
```

```python
manager_totalsales = merged_data.groupby(["manager"]).agg({"close_value":'sum'}) #summing up all the cost generated for specific managers
manager_totalsales
```

PROBLEM 2: Display 'Sales Agents' and 'Sales' for those sales where product sold at profit

```python
#solution code here
#merging the product and sales pipeline data 
merged_data2 = products.merge(sales_pipeline,on=["product"]) 
merged_data2
```

```python
# SPVSSPD = sales price vs sold price difference
merged_data2["SPVSSPD"] = merged_data2['sales_price'] - merged_data2['close_value'] 
merged_data2
```

Above is just a approach i tried where i went ahead with finding the difference but we Can't go with upper approach in this scenario as there are deal_stage where the sales team lost SPVSSPD won't be helpful in that case

```python
# RESULT as it was not specified in what all details should be present in sales with sales agent name i have printed all 
salesagent_sales = merged_data2[merged_data2['close_value']>merged_data2['sales_price']]
salesagent_sales
```

PROBLEM 3: Display the 'Opportunity ID' and 'Days Taken to Close', for opportunities those got closed within a month

```python
#Converting "Observation Date" into Datetime format
merged_data2["created_on"]=pd.to_datetime(merged_data2["created_on"])
merged_data2["close_date"]=pd.to_datetime(merged_data2["close_date"])
# days_since = Time taken to close the deal
merged_data2['days_since'] = merged_data2['close_date'] - merged_data2['created_on']
```

```python
merged_data2
```

```python
#RESULT 
merged_data2.loc[merged_data2['days_since'] <= '30 days', ['opportunity_id', 'days_since']]
```
